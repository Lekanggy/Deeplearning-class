```python
x = torch.arange(4.0)
x
```

```python
#Store the value to avoid memory leakage
x.requires_grad_(True)
x.grad
```

```python
# Define the equation
y = 2 * torch.dot(x,  x)
y
```

```python
#Calclate the gradient(Call the backward function)
y.backward()
x.grad
```

```python
#Calculate anotheer gradient
x.grad.zero_()
y = x.sum()
y.backward()
x.grad
```
