# Simple Single Layer Perceptron

```python
import numpy as np
```

```python
# Given data
x = np.array([1.0, 2.0])      # Inputs
y = 1                         # target
w = np.array([0.2, -0.5])     #weight
b = 0.1                       # bias
eta = 0.1                     #learning rate
```

```python
for i in range(1, 4):
  print(f"\nNumber of epoch: {i}")
  net_input = np.dot(w, x) + b
  y_pred = 1 if net_input > 0 else 0
  error = y - y_pred

  print(f"net_input = {net_input:.2f}")
  print(f"y_pred = {y_pred}")
  print(f"error = {error}")

  # Update weight and bias
  w += eta * error * x
  b += eta * error

  print(f"Updated weights: {w}")
  print(f"Updated bias: {b:.2f}")
```
